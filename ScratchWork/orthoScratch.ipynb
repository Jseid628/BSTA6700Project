```python
import torch
num_donors = 4
D = 2
Q = torch.eye(num_donors, D, dtype=torch.float64, requires_grad=True)  
print(Q) 
```

```python
print(Q.T@Q)
```

```python
torch.manual_seed(215)
import numpy as np
np.random.seed(42)


num_outcomes = 2
outcomes = torch.rand(num_outcomes, num_donors, dtype=torch.float64)
print(outcomes)
print("norm of outcomes:\n",torch.norm(outcomes))

from scipy.stats import ortho_group
Q = ortho_group.rvs(dim=4)
Q = torch.from_numpy(Q)
Q = Q[:,0:3]
print("Sliced Q:\n",Q)

print("Product:\n",outcomes@Q)
print("norm of transformed outcomes:\n",torch.norm(outcomes@Q))




print(Q[:,0:1].T@Q[:,0:1])
```
